# Pickle testing

This notebook should help with local testing of data summary and pickle creation, before integrating code changes into pipeline runs.

To get started, create the same conda environment as used by the pipeline step itself.
The step environment is declared in this repo, in `environment/environment.yml`.

From the command line:

```
git clone https://github.com/geffenlab/geffenlab-data-summary.git

cd geffenlab-data-summary

conda env create -f environment/environment.yml
```

Open this notebook in eg VS Code and select the appropriate Python kernel.
In VS Code this would be:
 - in the top-right corner, choose the Python kernel name
 - in the drop-down menu at the top, choose "Select Another Kernel..."
 - choose "Python Environments..." 
 - choose "geffenlab-data-summary"
 - VS Code might need to install the "ipykernel" package 

You should now be able to run this notebook.
